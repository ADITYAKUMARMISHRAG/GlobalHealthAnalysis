```python
gender_disparity = data.groupby(['Disease Category', 'Gender']).sum(numeric_only=True).reset_index()
```

```python
gender_diff = gender_disparity.pivot(index='Disease Category', columns='Gender', values=['Prevalence Rate (%)', 'Incidence Rate (%)', 'Mortality Rate (%)', 'DALYs'])
```

```python
gender_diff.columns = ['_'.join(col).strip() for col in gender_diff.columns.values]
```

```python
gender_diff = gender_diff.rename(columns={
    'Prevalence_Male': 'Prevalence_Male', 
    'Prevalence_Female': 'Prevalence_Female',
    'Incidence_Male': 'Incidence_Male',
    'Incidence_Female': 'Incidence_Female',
    'Mortality_Male': 'Mortality_Male',
    'Mortality_Female': 'Mortality_Female',
    'DALYs_Male': 'DALYs_Male',
    'DALYs_Female': 'DALYs_Female'
})
```

```python
gender_diff['DALYs_Difference'] = gender_diff['DALYs_Male'] - gender_diff['DALYs_Female']
```

```python
top_diseases = data.groupby('Disease Category')['DALYs'].sum().sort_values(ascending=False).head(5)
```

```python
# Plot 1: Gender disparities in DALYs
plt.figure(figsize=(10, 6))
gender_diff['DALYs_Difference'].sort_values().plot(kind='bar', color='teal')
plt.title('Gender Disparities in DALYs by Disease Category')
plt.ylabel('DALYs Difference (Male - Female)')
plt.xlabel('Disease Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('gender_disparities_dalys.png')  # Save plot as an image
plt.show()
```

```python
# Plot 2: Top disease categories by total DALYs
plt.figure(figsize=(10, 6))
top_diseases.plot(kind='bar', color='coral')
plt.title('Top Disease Categories by Total DALYs')
plt.ylabel('Total DALYs')
plt.xlabel('Disease Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('top_disease_categories_dalys.png')  # Save plot as an image
plt.show()
```

```python
# Plot 3: Line plot for DALYs trends (Line Plot)
plt.figure(figsize=(12, 6))
gender_diff[['DALYs_Male', 'DALYs_Female']].sort_values(by='DALYs_Male', ascending=False).plot(kind='line', marker='o')
plt.title('DALYs Trends for Male and Female Across Disease Categories')
plt.ylabel('DALYs')
plt.xlabel('Disease Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('dalys_trends_lineplot.png')  # Save plot as an image
plt.show()
```

```python
# Plot 4: Distribution of DALYs (Box Plot)
plt.figure(figsize=(8, 6))
sns.boxplot(data=gender_diff[['DALYs_Male', 'DALYs_Female']], palette='Set2')
plt.title('Distribution of DALYs for Male and Female')
plt.ylabel('DALYs')
plt.xlabel('Gender')
plt.xticks([0, 1], ['Male', 'Female'])
plt.tight_layout()
plt.savefig('dalys_distribution_boxplot.png')  # Save plot as an image
plt.show()
```

```python
dalys_distribution = data.groupby('Disease Category')['DALYs'].sum()
```

```python
# Pie Chart: Distribution of Total DALYs by Disease Category
plt.figure(figsize=(8, 8))
dalys_distribution.plot(kind='pie', autopct='%1.1f%%', startangle=90, cmap='tab20')
plt.title('Distribution of Total DALYs by Disease Category')
plt.ylabel('')  # Remove y-axis label for aesthetics
plt.tight_layout()
plt.savefig('dalys_distribution_pie_chart.png')  # Save pie chart as an image
plt.show()
```

```python
# Conclusion:
# 1. Gender disparities in DALYs are evident, with significant differences in disease impact between males and females. 
#    Certain diseases show a higher burden on males, while others affect females more prominently.
# 2. The top disease categories by DALYs highlight critical areas requiring focused health policies and interventions.
# 3. Correlation analysis between gender-specific metrics (Prevalence, Incidence, Mortality) indicates interconnected factors 
#    that can guide resource allocation and targeted strategies.
# 4. Visualizations provide a comprehensive understanding of trends and distributions, supporting data-driven decision-making in global health.
```

```python
data=data[['Country', 'Year', 
      #'Disease Name',
      'Disease Category',
       #'Prevalence Rate (%)',
      'Incidence Rate (%)',
      #'Mortality Rate (%)',
       #'Age Group',
      'Gender',
      #'Population Affected', 'Healthcare Access (%)',
       #'Doctors per 1000', 'Hospital Beds per 1000', 
      'Treatment Type',
       #'Average Treatment Cost (USD)', 'Availability of Vaccines/Treatment',
       #'Recovery Rate (%)', 'DALYs', 'Improvement in 5 Years (%)',
       #'Per Capita Income (USD)', 'Education Index', 'Urbanization Rate (%)'
     ]]
```

```python
dat2=data.set_index('Country')
```

```python
nat=input("please input the Country Name Here:")
```

```python
gend=input("please input gender here:")
```

```python
dattemp=dat2.loc[nat]
```

```python
NationGendData=dattemp[dattemp['Gender'].str.contains(gend)].sort_values(by='Incidence Rate (%)',ascending=False)
```

```python
NationGendData
```

```python
diseaseNationData=NationGendData.groupby('Disease Category')
```

```python
diseaseNationData['Incidence Rate (%)'].mean()
```

```python
y=diseaseNationData['Incidence Rate (%)'].mean().idxmax()
```

```python
filtered_data = NationGendData[NationGendData['Disease Category'] ==y]

# Step 3: Find the most sought-after treatment type for this disease category
most_sought_treatment = filtered_data['Treatment Type'].value_counts().idxmax()
```

```python
plt.figure(figsize=(16, 8))  
plt.title(f"Most Sought after Treatment Options for {gend}s Across Disease Categories in {nat}")

# Reduce the number of ticks if necessary and adjust label rotation and alignment
plt.xticks(rotation=45, ha='right', fontsize=12)

# Plot the data
diseaseNationData['Treatment Type'].value_counts().plot(kind="line", color='orange', linewidth=4, linestyle='--')

# Adjust the layout to prevent clipping
plt.tight_layout()  
plt.subplots_adjust(bottom=0.2)  # Increase bottom margin if labels are still getting cut off
plt.savefig('top_disease_categories_in_nation.png')  # Save plot as an image
# Show the plot
plt.show()
```

```python
print(f"Maximum occuring disease category among {gend}s in {nat} is {y} Thus the most sought after treatment type is {most_sought_treatment} "  )
```

```python
x=diseaseNationData['Incidence Rate (%)'].mean().idxmin()
```

```python
filtered_data2 = NationGendData[NationGendData['Disease Category'] ==x]

# Step 3: Find the most sought-after treatment type for this disease category
most_sought_treatment2 = filtered_data2['Treatment Type'].value_counts().idxmin()
```

```python
plt.figure(figsize=(12, 6))
plt.title(f"Trend Of Treatment Options Sought for {x} diseases in  {gend}s in {nat}")
plt.xticks(ticks=range(len(diseaseNationData['Disease Category'])),rotation=45, ha='right') 
plt.tight_layout()
filtered_data['Treatment Type'].value_counts().tail().plot(kind="line",color='green',linewidth=5, linestyle='--')
plt.savefig('top_disease_categories_in_nation2.png')  # Save plot as an image
plt.show()
```

```python
plt.figure(figsize=(8, 8))
filtered_data2['Treatment Type'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, cmap='tab20',pctdistance=0.85,labeldistance=1.05)
plt.title(f"Trend Of Treatment Options Sought for {x} diseases in  {gend}s in {nat}")
plt.ylabel('')  # Remove y-axis label for aesthetics

plt.tight_layout()
plt.savefig('treatment_pie_chart.png')  # Save pie chart as an image
plt.show()
```

```python
print(f"Minimum occuring disease category among {gend}s in {nat} is {x} Thus the least sought after treatment type is {most_sought_treatment2} "  )
```

```python
plt.figure(figsize=(12, 6),facecolor='orange')
plt.title(f"Incidence Rate Trends for diseases in {gend}s Across Disease Categories in {nat}")
plt.ylabel(' Incidence Rate')
plt.xticks(ticks=range(len(diseaseNationData['Disease Category'])),rotation=45, ha='right') 

plt.tight_layout()
diseaseNationData['Incidence Rate (%)'].mean().plot(kind="line",color='red', linewidth=3, linestyle='--')
plt.savefig('min_disease_categories_in_nation.png')  # Save plot as an image
plt.show()
```
